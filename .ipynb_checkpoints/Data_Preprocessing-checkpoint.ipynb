```python
import pandas as pd
import spacy
from spacy.lang import tl, en
from tglstemmer import stemmer
import simplemma
import string as s
import nltk
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from tensorflow.keras.layers import Dense, InputLayer
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.losses import BinaryCrossentropy
from tensorflow.keras import Sequential
```

```python
df = pd.read_csv("full.csv")
df.shape
```

```python
stopwords_tl = tl.stop_words.STOP_WORDS # Importing
stopwords_en = en.stop_words.STOP_WORDS
nltk.download('punkt_tab')
```

```python
stemmer_tl = stemmer
def word_proc(data):
    data = data.translate(str.maketrans('', '', s.punctuation))
    data = data.lower()
    stems = stemmer_tl.get_stems(data)
    data_stem = [simplemma.lemmatize(word, lang=('en'), greedy=True) for word in stems]
    data_sw_tl =[word for word in data_stem if word not in stopwords_tl]
    data_sw_en =[word for word in data_sw_tl if word not in stopwords_en]
    return " ".join(data_sw_en)
```

```python
# This function was created with the assistance of ChatGPT; this is used purely for preprocessing only. 
# The CSV file was generated to be used as a checkpoint as the preprocessing method I used took a long time to finish.
def word_proc_new(data):
    if not isinstance(data, str):
        return None  # Mark for removal later

    try:
        data = data.translate(str.maketrans('', '', s.punctuation))
        data = data.lower()

        stems = stemmer_tl.get_stems(data)
        if not isinstance(stems, list):
            return None  # Mark for removal

        data_stem = []
        for word in stems:
            if isinstance(word, str) and word:
                lemma = simplemma.lemmatize(word, lang='en', greedy=True)
                data_stem.append(lemma)

        data_sw_tl = [word for word in data_stem if word not in stopwords_tl]
        data_sw_en = [word for word in data_sw_tl if word not in stopwords_en]
        data_final = " ".join(data_sw_en)

        return data_final

    except Exception as e:
        # Optional: log or print for debugging
        print(f"Error processing: {data.index} | Exception: {e}")
        return None  # Mark for removal
```

```python
X = df
X_clean = X.article.apply(word_proc_new)
```

```python
samp_data = df['article'].loc[0]

data = samp_data.translate(str.maketrans('', '', s.punctuation))
data = data.lower()
stems = stemmer_tl.get_stems(data)
data_stem = [simplemma.lemmatize(word, lang=('en'), greedy=True) for word in stems]
data_sw_tl =[word for word in data_stem if word not in stopwords_tl]
data_sw_en =[word for word in data_sw_tl if word not in stopwords_en]
data_final = " ".join(data_sw_en)


data_final
```

```python
### Opening a checkpoint file made earlier
df_clean = pd.read_csv("full_clean_checkpoint.csv")
X_train, X_test, y_train, y_test = train_test_split(df_clean.article, df_clean.label, 
                                                     test_size=0.33, stratify=df_clean.label, 
                                                     random_state=42)
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

wordcloud = WordCloud(width=800, height=400, background_color='white').generate(df_clean.article[19])

plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')  
plt.title("Word Cloud fake news")
plt.show()
```

```python
vectorizer = TfidfVectorizer()
X_train_vec = vectorizer.fit_transform(X_train)
X_test_vec = vectorizer.transform(X_test)
```

```python
from tensorflow.keras.layers import Dense, InputLayer, Dropout
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.losses import BinaryCrossentropy
from tensorflow.keras import Sequential
from tensorflow.keras import regularizers
from sklearn.metrics import accuracy_score, roc_auc_score, RocCurveDisplay, classification_report
import numpy as np

model = Sequential()
model.add(InputLayer(shape=[X_train_vec.shape[1],]))
model.add(Dense(25, activation='relu', kernel_regularizer=regularizers.l2(0.001)))
model.add(Dropout(0.4))
model.add(Dense(1, activation='sigmoid'))
model.compile(optimizer=Adam(learning_rate=1e-4),loss=BinaryCrossentropy(), metrics=['accuracy'])
```

```python
model.fit(X_train_vec, y_train, validation_split=0.2, epochs = 300, shuffle=True)
```

```python
y_pred = model.predict(X_test_vec)
accuracy_score(y_test,np.round(y_pred))
```

```python
roc_auc_score(y_test, y_pred)
```

```python
RocCurveDisplay.from_predictions(y_test, y_pred)
```
